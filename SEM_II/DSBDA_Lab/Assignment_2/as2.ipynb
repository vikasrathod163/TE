```python
from scipy import stats
z = np.abs(stats.zscore(newdf['math score']))
threshold = 0.20
sample_outlier = np.where(z < threshold)
sample_outlier
```

```python
newnewdf = newdf

for i in sample_outlier:
    newnewdf.drop(i,inplace=True)

newnewdf
```

```python
newnewdf['math score'].plot(kind='hist')
newdf['log math'] = np.log10(newdf['math score'])
newdf['log math'].plot(kind='hist')
plt.show()
```
